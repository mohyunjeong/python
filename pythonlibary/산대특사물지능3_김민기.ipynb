## 문제 4.

```python
score_list = [["김미희", 89, 78, 99],
             ["서대희", 67, 88, 77],
             ["강예진", 100, 55, 96],
             ["차현석", 75, 100, 60]]
s_p = 0
s_j = 0
s_d = 0
print("="*45)
print("이름     Python  Java    DB   평균   학점")
print("="*45)
for s in range(len(score_list)):
    a_s = int(sum(score_list[s][1:])/3)
    s_p = s_p + score_list[s][1]
    s_j = s_j + score_list[s][2]
    s_d = s_d + score_list[s][3]
    
    if a_s >= 90 :
        print(f"{score_list[s][0]}     {score_list[s][1]}     {score_list[s][2]}     {score_list[s][3]}     {a_s}     A")
    elif a_s >= 80 :
        print(f"{score_list[s][0]}     {score_list[s][1]}     {score_list[s][2]}     {score_list[s][3]}     {a_s}     B")
    elif a_s >= 70 :
        print(f"{score_list[s][0]}     {score_list[s][1]}     {score_list[s][2]}     {score_list[s][3]}     {a_s}     C")
    elif a_s >= 60 :
        print(f"{score_list[s][0]}     {score_list[s][1]}     {score_list[s][2]}     {score_list[s][3]}     {a_s}     D")
    else:
        print(f"{score_list[s][0]}     {score_list[s][1]}     {score_list[s][2]}     {score_list[s][3]}     {a_s}     F")

print("="*45)
print(f"평균       {int(s_p/len(score_list))}     {int(s_j/len(score_list))}     {int(s_d/len(score_list))}")
```

## 문제 5.

```python
x = [2, 3, 4, 5, 6, 7, 8]
y1 = [5, 7, 3, 9, 1, 8, 6]
y2 = [9, 5, 7, 2, 3, 7, 1]

plt.plot(x,y1, label = "y1",c = "red")
plt.plot(x,y2, label = "y2",c = "blue")
plt.legend()
plt.grid()

plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

## 문제 6.

```python
ta_2017= pd.read_csv("Traffic_Accident_2017.csv", encoding = "euc-kr")
n_a = ta_2017.query("사고유형_대분류 == '차대사람' and 주야 == '야간' ")
n_a.shape # 조건 1. 964건

temp = n_a["요일"].value_counts()
temp.sort_index()

y = temp[["월", "화", "수", "목", "금", "토", "일"]]

x = range(7)
plt.bar(x,y)
plt.xticks(x, ["월", "화", "수", "목", "금", "토", "일"])
plt.ylim(0,200) # 조건 2. y 축 범위

plt.title("요일별 야간 차대사람 교통사고 건수")

plt.xlabel("요일")
plt.ylabel("사고건수")

plt.show()
```
