### 문제3. 아래와 같이 리스트를 만들고 요소의 짝수/홀수를 구분하여 개수를 세는 프로그램을 작성하시오. 

```python
list1 = [1, 50, 27, 19, 37, 33, 5, 3, 48]

even = odd = 0

for i in list1 :
    if (i % 2) == 0 :
        even += 1
    else :
        odd += 1
print(f'짝수 {even}개, 홀수{odd}개')
```

### 문제5. X, Y1, Y2 리스트 자료를 활용하여 [실행결과]와 같은 그래프를 출력하는 프로그램을 작성하시오. (단, 두 차트는 빨간색과 파란색을 통해 구별 하시오.)

```python
import matplotlib.pyplot as plt

x = [2, 3, 4, 5, 6, 7, 8]
y1 = [5, 7, 3, 9, 1, 8, 6]
y2 = [9, 5, 7, 2, 3, 7, 1]

plt.plot(x, y1, label = 'y1', c = 'r')
plt.plot(x, y2, label = 'y2', c = 'b')

plt.legend()

plt.title('차트', loc = 'right')

plt.xlabel('x')
plt.ylabel('y')

plt.show()
```
