4. üü¶ Ï¥àÎ•º ÏûÖÎ†• Î∞õÏïÑÏÑú "00ÏãúÍ∞Ñ 00Î∂Ñ 00Ï¥à" ÌòïÌÉúÎ°ú Ï∂úÎ†•ÌïòÏãúÏò§ üü¶


```python
# üü™ Î∞©Î≤ï1 üü™
time = int(input("ÏãúÍ∞Ñ ÏûÖÎ†• >> "))

# 1ÏãúÍ∞ÑÏùÄ 3600Ï¥àÏù¥Í∏∞ ÎïåÎ¨∏Ïóê
hour = time // 3600
minute = (time % 3600) // 60
second = time % 60

print(f"\n{hour}ÏãúÍ∞Ñ {minute}Î∂Ñ {second}Ï¥à")
```

```python
# üü™ Î∞©Î≤ï2 üü™
time = int(input("ÏãúÍ∞Ñ ÏûÖÎ†• >> "))

hour = time // 3600
minute = (time // 60) - (hour * 60) # ÏãúÍ∞ÑÏùÑ Î∂ÑÏúºÎ°ú Î∞îÍøî Í≥ÑÏÇ∞
second = time - ((minute * 60) + (hour * 3600)) # ÏãúÍ∞ÑÍ≥º Î∂ÑÏùÑ Ï¥àÎ°ú Î∞îÍøî Í≥ÑÏÇ∞

print(f"\n{hour}ÏãúÍ∞Ñ {minute}Î∂Ñ {second}Ï¥à")
```

### üüß ÎÖºÎ¶¨ Ïó∞ÏÇ∞Ïûê üüß
- True, FalseÏùò ÎÖºÎ¶¨Í∞í(booleanÍ∞í)ÏùÑ Î∞òÌôòÌïòÎäî Ïó∞ÏÇ∞Ïûê

- üü® not Ïó∞ÏÇ∞Ïûê üü®

```python
a = 3
b = 7

print(f'a < b : {a < b}')
print(f'not a < b : {not a < b}') # not : Î∂ÄÏ†ïÏó∞ÏÇ∞Ïûê -> True or FalseÏùò Î∞òÎåÄ Í∞í Ï∂úÎ†•
```

- üü® and Ïó∞ÏÇ∞Ïûê : Îëê Ï°∞Í±¥Ïù¥ Î™®Îëê Ï∞∏(True)Ïùº Í≤ΩÏö∞ÏóêÎßå Ï∞∏ÏùÑ Î∞òÌôò üü®

```python
print(3 > 5) # --> False
print(10 == 20) # --> False

# üü© Îëò Îã§ False üü©

print(3 > 5 and 10 == 20) # False
```

```python
print(3 < 5) # --> True
print(10 == 20) # --> False

# üü©ÌïòÎÇòÎßå True üü©

print(3 < 5 and 10 == 20) # False
```

```python
print(3 < 5) # --> True
print(10 != 20) # --> True

# üü© Îëò Îã§ True üü©

print(3 < 5 and 10 != 20) # True
```

- üü® or Ïó∞ÏÇ∞Ïûê : Îëê Ï°∞Í±¥ Ï§ë ÌïòÎÇòÎùºÎèÑ Ï∞∏(True)Ïù¥Î©¥ Ï∞∏ Î∞òÌôò üü®

```python
print(3 < 5) # --> True
print(10 == 20) # --> False

# üü©ÌïòÎÇòÎßå True üü©

print(3 < 5 or 10 == 20) # True
```

```python
print(3 < 5) # --> True
print(10 != 20) # --> True

# üü© Îëò Îã§ True üü©

print(3 < 5 or 10 != 20) # True
```
